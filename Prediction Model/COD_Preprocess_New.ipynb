#Convert float categorical columns to proper categorical type and encode 'Sex'

```python

# First encode Sex (M -> 1, F -> 0)
df['Sex'] = df['Sex'].map({'M': 1, 'F': 0})

# Define all categorical columns including Sex
categorical_columns = [
    'Sex',
    'Nourished',
    'Age- Old',
    'Chest cavity-Free air or adhesions',
    'Heart-Size in configuration',
    'Heart-Injuries',
    'Heart-Normal myocardium',
    'Heart-Recent ischaemic changes',
    'Heart-Myocardial fibrosis present in',
    'Heart-Concentric hypertrophy detected',
    'Heart-Valves normal',
    'Blood Vessels-Coronary arteries',
    'Blood Vessels-Free of Stenosis',
    'Blood Vessels-Coronary artery condition',
    'Blood Vessels-clacified and stenosed with patchy atheromatous plaques',
    'Aorta-Aorta condition',
    'Aorta-Aorta Injuries',
    'Aorta-Atheromatous plaques present in the aorta',
    'Cause of Death'
]

# Handle missing values and convert to integer for all categorical columns
for col in categorical_columns:
    if col in df.columns:
        # Fill NaN values with mode
        mode_val = df[col].mode().iloc[0] if not df[col].mode().empty else None
        df[col] = df[col].fillna(mode_val)
        # Convert to integer first to remove decimal points
        df[col] = df[col].astype(int)
        # Then convert to categorical
        df[col] = pd.Categorical(df[col])
```

```python
# Define continuous columns
continuous_columns = [
    'Pericardium-pericardial fluid - in ml',
    'Heart-Size',
    'Blood Vessels-Left anterior descending artery',
    'Blood Vessels-right coronary artery',
    'Blood Vessels-Left circumflex artery'
]

# Fill missing values in continuous columns
for col in continuous_columns:
    if col in df.columns:
        mode_val = df[col].mode().iloc[0] if not df[col].mode().empty else None
        df[col] = df[col].fillna(mode_val)
```

# Separate the DataFrame based on the condition 'cause of death == 1'

```python
df_cause_of_death_1 = df[df['Cause of Death'] == 1].copy()

# Fill missing values in continuous columns with the mean for 'cause of death == 1'
for col in ['Blood Vessels-Left anterior descending artery', 'Blood Vessels-right coronary artery', 'Blood Vessels-Left circumflex artery']:
    if not df_cause_of_death_1[col].empty:
        mean_value = df_cause_of_death_1[col].mean()
        df_cause_of_death_1[col] = df_cause_of_death_1[col].fillna(mean_value)

# Update the original DataFrame with the filled values
df.update(df_cause_of_death_1)
```

# Separate the DataFrame based on the condition 'cause of death == 0'

```python
# Separate the DataFrame based on the condition 'cause of death == 0'
df_cause_of_death_0 = df[df['Cause of Death'] == 0].copy()

# Fill missing values in continuous columns with 0 for 'cause of death == 0'
for col in ['Blood Vessels-Left anterior descending artery', 'Blood Vessels-right coronary artery', 'Blood Vessels-Left circumflex artery']:
    if not df_cause_of_death_0[col].empty:
        df_cause_of_death_0[col] = df_cause_of_death_0[col].fillna(0)

# Update the original DataFrame with the filled values for 'cause of death == 0'
df.update(df_cause_of_death_0)
```

#Handling Corelating features

```python
from sklearn.decomposition import PCA

# Extract the correlated columns
correlated_features = ['Blood Vessels-Left anterior descending artery',
                      'Blood Vessels-right coronary artery',
                      'Blood Vessels-Left circumflex artery']

# Create PCA object
pca = PCA(n_components=1)

# Fit and transform correlated features in training data
correlated_features_train = x_train[correlated_features].values
pca_result_train = pca.fit_transform(correlated_features_train)

# Transform test data
correlated_features_test = x_test[correlated_features].values
pca_result_test = pca.transform(correlated_features_test)

# Drop original correlated columns
x_train_modified = x_train.drop(columns=correlated_features)
x_test_modified = x_test.drop(columns=correlated_features)


# Add PCA component back as a new feature
x_train_modified['vessels_pca'] = pca_result_train
x_test_modified['vessels_pca'] = pca_result_test
```

#Building the model using random forrest

```python
# Import Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Initialize the classifier
classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model - note that we use y_train, not y_train_modified
classifier.fit(x_train_modified, y_train)

# Make predictions
pred_classifier = classifier.predict(x_test_modified)

# Calculate accuracy - use the modified datasets
print("Acc on training data: {:.3f}".format(classifier.score(x_train_modified, y_train)))
print("Acc on test data: {:.3f}".format(classifier.score(x_test_modified, y_test)))
```

#Randomized Search Cv

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt','log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 1000,10)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10,14]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4,6,8]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion':['entropy','gini']}
print(random_grid)
```

```python
rf=RandomForestClassifier()
rf_randomcv=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_iter=20,cv=2,verbose=2,
                               random_state=100,n_jobs=-1)
### fit the randomized model
rf_randomcv.fit(x_train_modified,y_train)
```

#Training with Naive Baiyes

```python
# Import necessary libraries for Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train_modified)
X_test_scaled = scaler.transform(x_test_modified)

# Initialize and train the Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)
```

```python
# Calculate training accuracy
train_accuracy = nb_model.score(X_train_scaled, y_train)

# Calculate test accuracy
test_accuracy = nb_model.score(X_test_scaled, y_test)

# Print results
print("Training Accuracy: {:.4f}".format(train_accuracy))
print("Test Accuracy: {:.4f}".format(test_accuracy))
```

```python
# Import necessary libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import numpy as np

# First, let's do RandomizedSearchCV
# Define the parameter distributions for GaussianNB
param_dist = {
    'var_smoothing': np.logspace(-10, -8, 100)  # This is the main hyperparameter for GaussianNB
}

# Initialize GaussianNB
nb = GaussianNB()

# Initialize RandomizedSearchCV
nb_randomcv = RandomizedSearchCV(
    estimator=nb,
    param_distributions=param_dist,
    n_iter=20,
    cv=2,
    verbose=2,
    random_state=100,
    n_jobs=-1
)

# Fit RandomizedSearchCV
nb_randomcv.fit(x_train_modified, y_train)

# Print best parameters and score
print("Best parameters found by RandomizedSearchCV:")
print(nb_randomcv.best_params_)
```

```python
# Now let's do GridSearchCV with a finer search around the best parameter
best_var_smoothing = nb_randomcv.best_params_['var_smoothing']
param_grid = {
    'var_smoothing': np.logspace(
        np.log10(best_var_smoothing) - 1,
        np.log10(best_var_smoothing) + 1,
        20
    )
}

# Initialize GridSearchCV
nb_grid = GridSearchCV(
    estimator=GaussianNB(),
    param_grid=param_grid,
    cv=3,
    verbose=2,
    n_jobs=-1
)

# Fit GridSearchCV
nb_grid.fit(x_train_modified, y_train)
```

```python
# Get the best model
best_nb = nb_grid.best_estimator_

# Make predictions
y_pred = best_nb.predict(x_test_modified)

# Print evaluation metrics
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Calculate and print training vs test accuracy
train_accuracy = best_nb.score(x_train_modified, y_train)
test_accuracy = accuracy_score(y_test, y_pred)

print("\nTraining Accuracy: {:.4f}".format(train_accuracy))
print("Test Accuracy: {:.4f}".format(test_accuracy))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Get probability predictions for Naive Bayes
nb_probs = best_nb.predict_proba(x_test_modified)[:, 1]

# Calculate ROC curve
fpr, tpr, _ = roc_curve(y_test, nb_probs)

# Calculate AUC
nb_auc = auc(fpr, tpr)

# Create ROC plot
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'Naive Bayes (AUC = {nb_auc:.3f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Customize the plot
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Naive Bayes')
plt.legend(loc='lower right')
plt.grid(True)

# Show the plot
plt.show()

# Print AUC score
print(f'Naive Bayes AUC: {nb_auc:.3f}')
```
